1. Determine the cost of some number of pounds of cement. Assume there is a sales tax that applies.
  

```python
# Exercise 1 - Part 1
# Determine the cost of some number of pounds of cement. Assume there is a sales tax that applies.

#1 Define variables
pounds_of_cement = 45  # Example: 100 pounds of cement
cost_per_pound = 2.50   # Example: $2.50 per pound
sales_tax_rate = 0.06   # Example: 6% sales tax

#2 Calculate costs
total_cost_before_tax = pounds_of_cement * cost_per_pound
sales_tax_amount = total_cost_before_tax * sales_tax_rate
total_cost_after_tax = total_cost_before_tax + sales_tax_amount

#3 You can print the result to verify
#4 print(f"Total cost of {pounds_of_cement} lbs of cement: ${total_cost_after_tax:.2f}")
```
